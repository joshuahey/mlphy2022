```python
def spectrogram_gen(input):
    spectrogram = tf.signal.stft(input, frame_length=250, frame_step=125)
    spectrogram = tf.abs(spectrogram)
    spectrogram = tf.expand_dims(spectrogram, axis=2)
    return spectrogram
```
